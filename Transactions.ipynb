```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop('Product_Code', axis=1)
data.head()
```

```python
X = data.iloc[:,0:52]
X_normalized = data.iloc[:,53:106]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
kmeans_norm = KMeans(n_clusters=2, random_state=0).fit(X_normalized)
```

```python
kmeans.labels_
```

```python
kmeans_norm.labels_
```
